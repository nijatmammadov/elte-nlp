```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

```python
import os
import random
import shutil

# Paths
base_path = '/content/drive/MyDrive/nlp_project_elte/od/od/train'
images_path = os.path.join(base_path, 'images')
labels_path = os.path.join(base_path, 'labels')

# Output folders
splits = ['train', 'val', 'test']
for split in splits:
    os.makedirs(os.path.join(base_path, f'images/{split}'), exist_ok=True)
    os.makedirs(os.path.join(base_path, f'labels/{split}'), exist_ok=True)

# Parameters
val_ratio = 0.2
test_ratio = 0.1

# Collect and shuffle image filenames
all_images = [f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png'))]
random.shuffle(all_images)

# Split
total = len(all_images)
val_count = int(total * val_ratio)
test_count = int(total * test_ratio)
train_count = total - val_count - test_count

val_files = all_images[:val_count]
test_files = all_images[val_count:val_count + test_count]
train_files = all_images[val_count + test_count:]

# Helper to move files
def move_files(file_list, split):
    for file in file_list:
        image_src = os.path.join(images_path, file)
        label_src = os.path.join(labels_path, file.rsplit('.', 1)[0] + '.txt')

        image_dst = os.path.join(base_path, f'images/{split}', file)
        label_dst = os.path.join(base_path, f'labels/{split}', file.rsplit('.', 1)[0] + '.txt')

        shutil.copy(image_src, image_dst)
        if os.path.exists(label_src):
            shutil.copy(label_src, label_dst)

# Move files
move_files(train_files, 'train')
move_files(val_files, 'val')
move_files(test_files, 'test')

print(f"Split complete: {len(train_files)} train, {len(val_files)} val, {len(test_files)} test.")
```

```python
model.train(data='/content/drive/MyDrive/nlp_project_elte/od/od/data.yaml',  # Path to YAML config
            epochs=50,                  # Number of epochs
            imgsz=640,                  # Image size
            batch=16,                   # Batch size
            device=0)
```

```python
results = model.val()
print(results)
```
